```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)

IQR = Q3-Q1

upper_bound = Q3+1.5*(IQR)
lower_bound = Q1-1.5*(IQR)
print(upper_bound , lower_bound)
```

```python
data = data[~((data>upper_bound))|((data<lower_bound))]
```
